The goal of this project is to look at a few different Hip Hop albums and compare some of the audio features that I think would correlate to a more popular song, or a "hit" song if you will.My plan is to grab all of the tracks from 3 albums from some on the biggest artists out right now. My criteria was albums that came out this year. This was because the popularity key from Spotify counts recent listens, not overall listens. For example, a song that came out and was a number one song 10 years ago may have a lower popularity then a somewhat new popular song because of how recent the plays are. I had to take this into account when choosing these albums. Spotify has quite a few differnet audio features, so I had to go look through the API and see what all of them actually mean. I decided on 4 features that I though might actually have an effect on whether a song is popular or not.

To start off the report, I imported anything that I might be needing in order to complete this project, as well as getting all of the spotify session stuff together.

```python
import json
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import base64
import requests
```

```python
keys = pd.read_csv("Spotify-Keys_10-18-22.txt", header = 0, sep = ",")
appid = keys['Client_ID'][0]
appsecret = keys['Client_Secret'][0]
client_cred = base64.b64encode(str(appid + ":" + appsecret).encode('ascii'))
header = {'Authorization': 'Basic {}'.format(client_cred.decode('ascii'))}
payload = {'grant_type' : 'client_credentials'}
access_token_url = "https://accounts.spotify.com/api/token"
response = requests.post(access_token_url, headers = header, data = payload)
session_token = response.json()['access_token']
session_header = {'Authorization': 'Bearer {}'.format(response.json()['access_token'])}
```

Next come the graphs. I am going to throw together some scatter plots because I feel like this will be the easiest way to read and understand the data. Each point will be a song which will make a trend easier to see.

```python
fig1, ax1 = plt.subplots(2,2,layout = 'constrained')
ax1[0,0].scatter(f_days_df['popularity'], f_days_df['danceability'], alpha = 0.5, color = 'red', label = 'Vinyl Days')
ax1[0,0].scatter(f_loss_df['popularity'], f_loss_df['danceability'], alpha = 0.5, color = 'blue', label = 'Her Loss')
ax1[0,0].scatter(f_king_df['popularity'], f_king_df['danceability'], alpha = 0.5, color = 'green', label = 'Kings Disease')
ax1[0,0].set_ylabel('Danceability')
ax1[0,0].set_xlabel('Popularity')
#
ax1[0,1].scatter(f_days_df['popularity'], f_days_df['valence'], alpha = 0.5, color = 'red', label = 'Vinyl Days')
ax1[0,1].scatter(f_loss_df['popularity'], f_loss_df['valence'], alpha = 0.5, color = 'blue', label = 'Her Loss')
ax1[0,1].scatter(f_king_df['popularity'], f_king_df['valence'], alpha = 0.5, color = 'green', label = 'Kings Disease')
ax1[0,1].set_ylabel('Valence')
ax1[0,1].set_xlabel('Popularity')
#
ax1[1,0].scatter(f_days_df['popularity'], f_days_df['energy'], alpha = 0.5, color = 'red', label = 'Vinyl Days')
ax1[1,0].scatter(f_loss_df['popularity'], f_loss_df['energy'], alpha = 0.5, color = 'blue', label = 'Her Loss')
ax1[1,0].scatter(f_king_df['popularity'], f_king_df['energy'], alpha = 0.5, color = 'green', label = 'Kings Disease')
ax1[1,0].set_ylabel('Energy')
ax1[1,0].set_xlabel('Popularity')
#
ax1[1,1].scatter(f_days_df['popularity'], f_days_df['duration_ms'], alpha = 0.5, color = 'red', label = 'Vinyl Days')
ax1[1,1].scatter(f_loss_df['popularity'], f_loss_df['duration_ms'], alpha = 0.5, color = 'blue', label = 'Her Loss')
ax1[1,1].scatter(f_king_df['popularity'], f_king_df['duration_ms'], alpha = 0.5, color = 'green', label = 'Kings Disease')
ax1[1,1].set_ylabel('Duration')
ax1[1,1].set_xlabel('Popularity')

ax1[0,0].legend()
plt.show()
```

After looking at these graphs for a little while, I do notice a few trends that are interesting. Danceability is all over the place. From what it looks like, there might be a slight upwards trend but the data is too all over the place to take any stock in that point. This was surprising to me because I really thought danceability was going to have the most effect on whether a song was popular or not, so I was wrong there. Valance on the other hand does seem so show a trend. From what it looks like, the lower a songs valence is, the higher a chance of its popularity. Energy also seems to follow that same behavior as valance. Duration on the other hand is a bit tricky. It seems that if your song is less than 1.7 minutes, there is a negative connotation. The being said, it looks like the sweet spot for a hip hop track is right around 3.3 minutes give or take. There are a few conclusion that I can take away from this data. If you were an up and coming artist in hip hop in todays game, I would say 3 things; 1: Make a song that is around 2.5-3.5 minutes long, 2:Lower Valence seems to be better, and 3: Lower energy seems to be better as well. That being said, this data is very incomplete and could use a lot more work in order to get some more concrete evidence of any correlations. The next steps should be to test with more albums and more artists. That could do one of two things, either outline a correlation even more, or completely disprove what was shown here. Either way, it is important that more is done before any conclusions are made.
